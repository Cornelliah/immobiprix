```python
import requests
import pandas as pd
import time 

adresse = "saint Blaise, Paris 20ème arrondissement (75020)"

url = "https://api-adresse.data.gouv.fr/search/"



def get_long_lat (adresse) : 
    time.sleep(0.001)
    params = {
    "q": adresse,
    "limit": 1}
    
    
    
    
    try :
        response = requests.get(url, params=params)
        data = response.json()
        if data["features"]:
            resultat = data["features"][0]
    
            return(resultat["geometry"]["coordinates"])
    
        else:
            return([])
        
    except :
        return([])
```

```python
get_long_lat("190 bis, boulevard de charonne, 75020")
```
